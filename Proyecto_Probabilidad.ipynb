### Prueba de Hipotesis

```python
loser_name = set(atptotal['Loser'].unique())
```

```python
winner_name = set(atptotal['Winner'].unique())
```

```python
total_match = winner_name.union(loser_name)
total_match
```

```python
tasa_wins={}

for jugador in total_match:
  novak = atptotal[(atptotal['Winner'] == jugador)  | (atptotal['Loser'] == jugador)]
  novakwin = novak[novak['Winner'] == jugador]
  nole = novak.shape[0]
  nolewin = novakwin.shape[0]
  tasa_wins[jugador] = nolewin/nole
  


```

```python
tasa_total = pd.DataFrame(tasa_wins.items(), columns=('Nombre', 'Tasa_wins'))
```

```python
from statsmodels.stats.weightstats import ztest
```

```python
ztest(tasa_total['Tasa_wins'], value=0.83, alternative='larger')
```

```python
def wins_año(date1, date2):
  jugador = atptotal[(atptotal['Winner'] == 'Nadal R.')  | (atptotal['Loser'] == 'Nadal R.')]
  fecha = jugador[(jugador['Date'] >= str(date1)) & (jugador['Date'] <= str(date2))]
  wins = fecha[fecha['Winner'] == 'Nadal R.']
  return wins.shape[0]/fecha.shape[0]
```
